```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("nld_train.csv")
```

```python
df.head()
```

```python
df.dropna()
```

```python
df['new'] = np.where(df['normal'].str.contains('normal'), 1, 0)
```

```python
df.dropna()
```

```python
df=df.drop(['normal'], axis=1)
```

```python
#df.info()
df.nunique()
```

```python
df=df.drop(['0.15'], axis=1)
df=df.drop(['0'], axis=1)
```

```python
target = ['new']
target[0]
```

```python
categorical = ['tcp','SF']
```

```python
df['ftp_data']=df['ftp_data'].astype('category')
df['ftp_data'] = df['ftp_data'].cat.codes
df['ftp_data']
```

```python
numerical = [col for col in df.columns 
            if col not in target+categorical]
```

```python
df = pd.get_dummies(data=df, columns=categorical, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to numerical columns
scaled_numerical = scaler.fit_transform(df[numerical])

# Build a DataFrame
scaled_numerical = pd.DataFrame(scaled_numerical, columns=numerical)
```

```python
df = df.drop(columns=numerical, axis=1)
```

```python

# Merge the non-numerical with the scaled numerical data
X = df.merge(right=scaled_numerical,
how='left',
left_index=True,
right_index=True
)
```

```python
X = X.drop(['new'], axis=1)
```

```python
Y = df.new
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                 X, Y, test_size=0.4, random_state=0)
```

```python
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
