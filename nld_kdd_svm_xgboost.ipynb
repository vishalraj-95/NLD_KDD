```python
import pandas as pd
import numpy as np
```

```python
#loading the train data set
df=pd.read_csv("nld_train.csv")
```

```python
df.head()
```

```python
#dropping the row having null value
df.dropna()
```

```python
#changing the label of independent feature
df['new'] = np.where(df['normal'].str.contains('normal'), 1, 0)
```

```python
df.dropna()
```

```python
df=df.drop(['normal'], axis=1)
```

```python
#df.info()
df.nunique()
```

```python
#dropping the axis with no value
df=df.drop(['0.15'], axis=1)
df=df.drop(['0'], axis=1)
```

```python
target = ['new']
target[0]
```

```python
categorical = ['tcp','SF']
```

```python
#label encoding
df['ftp_data']=df['ftp_data'].astype('category')
df['ftp_data'] = df['ftp_data'].cat.codes
df['ftp_data']
```

```python
numerical = [col for col in df.columns 
            if col not in target+categorical]
```

```python
#one hot encoding
df = pd.get_dummies(data=df, columns=categorical, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to numerical columns
scaled_numerical = scaler.fit_transform(df[numerical])

# Build a DataFrame
scaled_numerical = pd.DataFrame(scaled_numerical, columns=numerical)
```

```python
df = df.drop(columns=numerical, axis=1)
```

```python

# Merge the non-numerical with the scaled numerical data
X = df.merge(right=scaled_numerical,
how='left',
left_index=True,
right_index=True
)
```

```python
#independent feature
X = X.drop(['new'], axis=1)
```

```python
# dependent feature
Y = df.new
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                 X, Y, test_size=0.4, random_state=0)
```

```python
#defining the svm model
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
```

```python
#cross validation score
clf.score(X_test, y_test)
```

```python
#test data
df_t=pd.read_csv("nld_test.csv")
```

```python
df_t
```

```python
#df_t.info()
df_t.nunique()
```

```python
df_t['new'] = np.where(df_t['normal'].str.contains('normal'), 1, 0)
```

```python
df_t.dropna()
```

```python
df_t.nunique()
```

```python
df_t=df_t.drop(['normal','0.15','0'], axis=1)
df_t
```

```python
target_t = ['new']
target_t[0]
```

```python
categorical_t =['tcp','SF']
categorical_t
```

```python
df_t['ftp_data']=df_t['ftp_data'].astype('category')
df_t['ftp_data'] = df_t['ftp_data'].cat.codes
df_t['ftp_data']
```

```python
numerical_t = [col for col in df_t.columns 
            if col not in target_t+categorical_t]
```

```python
df_t = pd.get_dummies(data=df_t, columns=categorical_t, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to numerical columns
scaled_numerical = scaler.fit_transform(df_t[numerical_t])

# Build a DataFrame
scaled_numerical = pd.DataFrame(scaled_numerical, columns=numerical_t)
```

```python
df_t = df_t.drop(columns=numerical_t, axis=1)
df_t
```

```python
# Merge the non-numerical with the scaled numerical data
X_t = df_t.merge(right=scaled_numerical,
how='left',
left_index=True,
right_index=True
)
```

```python
X_t = X_t.drop(['new'], axis=1)
```

```python
Y_t = df_t.new
Y_t
```

```python
from sklearn.svm import SVC
# 2. Initialize the model
svm_clf=SVC(gamma="auto")
```

```python
# Fit the model on the training data
svmmodel = svm_clf.fit(X, Y)
```

```python
X_t.info()
```

```python
# 4. Predict values on the testing data
pred_Y = svmmodel.predict(X_t)
```

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
accuracy_score(Y_t,pred_Y )
```

```python
# average precision score
from sklearn.metrics import average_precision_score
average_precision = average_precision_score(Y_t,pred_Y)

print('Average precision-recall score: {0:0.2f}'.format(
      average_precision))
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
import matplotlib.pyplot as plt

disp = plot_precision_recall_curve(svm_clf, X_t, Y_t)
disp.ax_.set_title('2-class Precision-Recall curve: '
                   'AP={0:0.2f}'.format(average_precision))
```
